```python
import os
import pandas as pd
from ast import literal_eval
from collections import Counter
from datetime import datetime
```

```python
DATA_DIR = os.getenv("DATA_DIR")
content_path = os.path.join(DATA_DIR, "preprocessed_content_store_en.csv.gz")
df = pd.read_csv(content_path, compression="gzip")
```

```python
df.shape
```

```python
df['taxons'] = df['taxons'].map(lambda x: literal_eval(x) if not isinstance(x,float) else [])
```

```python
df['num_taxons'] = df['taxons'].map(len)
df['num_taxons'].describe()
```

```python
#### none because world taxons are unrooted
df[df['num_taxons']==df['num_taxons'].max()].taxons.values
```

```python
df[df['num_taxons']>10].iloc[0].taxons
```

```python
row_list = []

taxon_col = df.columns.get_loc('taxons')
cid_col = df.columns.get_loc('content_id')
base_path_col = df.columns.get_loc('base_path')
print(datetime.now().strftime("%H:%M:%S"))

for tup in df.itertuples(index=False):
    taxon_track = []
    if len(tup[taxon_col]) >0:
        for taxon in tup[taxon_col]:
            if taxon!=None:
#                 print(taxon)
                if taxon['title'] not in taxon_track:
                    taxon_track.append(taxon['title'])
                    row_dict = {'content_id': tup[cid_col],
                                'base_path': tup[base_path_col],
                                'taxon_title': taxon['title'],
                                'taxon_base_path':  taxon['base_path'],
                                'taxon_content_id': taxon['content_id']
                         }
                    row_list.append(row_dict)
                else:
                    taxon_track.append(taxon['title'])
print(datetime.now().strftime("%H:%M:%S"))                
df_cid_per_tax = pd.DataFrame(row_list)        
```
